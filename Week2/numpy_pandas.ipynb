```python
# BTVN 1 
# cho ma trận bằng [[99,99,99],
#                   [99,99,99],
#                   [99,99,99]]

# hãy tưởng tượng ô 99 trống
# giả sử 1 là x 
# giả sử 0 là o 
# lấy đầu vào từ phía X và phái O luân phiên 
# Nếu đầu vào từ phía x là (0,0)
# thì ma trận sẽ trở thành 
#                  [[x,99,99],
#                   [99,99,99],
#                   [99,99,99]]

# Nếu đầu vào từ phía o là (0,0)
# Thông báo họ nhập lại 
# nếu không thì điền vào ma trận

# Thử thách : Nếu người đầu tiên có 3 phần tử liên tục thì dừng trò chơi

import numpy as np

def check_win(board, player):
    # Check rows
    for i in range(3):
        if np.all(board[i, :] == player):
            return True
    # Check columns
    for i in range(3):
        if np.all(board[:, i] == player):
            return True
    # Check diagonals
    if np.all(np.diag(board) == player):
        return True
    if np.all(np.diag(np.fliplr(board)) == player):
        return True
    return False

# Initialize board
board = np.full((3, 3), 99)
current_player = 1  # 1 for X, 0 for O

while True:
    print("Bảng hiện tại:")
    print(board)
    print(f"Player {'X' if current_player == 1 else 'O'}'s turn")
    
    # Get input
    try:
        row, col = map(int, input("Nhập số hàng và số cột (0-2, cách nhau bằng dấu cách):").split())
    except ValueError:
        print("Nhập liệu không hợp lệ. Vui lòng nhập hai số cách nhau bằng dấu cách.")
        continue
    
    # Check bounds
    if not (0 <= row < 3 and 0 <= col < 3):
        print("Vị trí nằm ngoài phạm vi. Thử lại.")
        continue
    
    # Special rule for O
    if current_player == 0 and (row, col) == (0, 0):
        print("O không thể chơi tại (0,0). Thử lại.")
        continue
    
    # Check if position is empty
    if board[row, col] != 99:
        print("Vị trí đã có người đánh. Vui lòng thử lại.")
        continue
    
    # Place the mark
    board[row, col] = current_player
    
    # Check for win
    if check_win(board, current_player):
        print("Bảng hiện tại:")
        print(board)
        print(f"Player {'X' if current_player == 1 else 'O'} wins!")
        break
    
    # Switch player
    current_player = 1 - current_player
```

```python
#BTVN 4 
# tạo 1 mảng numpy với kích thước 150x5 , Giả sử mảng này chưa 150 mẫu
# dữ liệu của sinh viên bao gồm chiều cao, cân nặng, tuổi , lương, điểm trung bình
# tách mảng x thành 2 dựa trên 4 cột đầu tiên thành 1 biến có tên là x và cột cuối cùng thành y

# tách x thành x_train và x_test trong đó x_train chưa 70% dữ liệu
# và tách y thành y_train và y_test , trong đó y_train chứa 70%
# dữ liệu

# Tạo 10 tập dữ liệu x_train không chồng chéo nhau 

import numpy as np

np.random.seed(42)  

data = np.random.rand(150, 5)

data[:, 0] = data[:, 0] * 50 + 150  # Chiều cao
data[:, 1] = data[:, 1] * 60 + 40   # Cân nặng
data[:, 2] = data[:, 2] * 7 + 18    # Tuổi
data[:, 3] = data[:, 3] * 100000    # Lương
data[:, 4] = data[:, 4] * 4         # GPA

print("Kích thước dữ liệu:", data.shape)

X = data[:, :4]
y = data[:, 4]

print("Kích thước X:", X.shape)
print("Kích thước y:", y.shape)

train_size = int(0.7 * 150)
X_train = X[:train_size]
X_test = X[train_size:]
y_train = y[:train_size]
y_test = y[train_size:]

print("Kích thước X_train:", X_train.shape)
print("Kích thước X_test:", X_test.shape)
print("Kích thước y_train:", y_train.shape)
print("Kích thước y_test:", y_test.shape)

indices = np.arange(len(X_train))
np.random.shuffle(indices)

splits = np.array_split(indices, 10)

x_trains = [X_train[split] for split in splits]
y_trains = [y_train[split] for split in splits]

print("Số lượng tập x_train:", len(x_trains))
for i, xt in enumerate(x_trains):
    print(f"Kích thước x_train[{i}]:", xt.shape)
```
