```python
# BTVN

# Cho y = 5x^5 + 6x^3 -3x + 1
# Cho biết độ dốc của đa thức trên ở điểm nào

#BTVN 2:
# Tạo một tensor ban đầu có giá trị là 2. Định nghĩa hàm số và tính ra gradient
# y = x^3 + 2x^2 + 5x + 1
# Hãy tính dy/dx tại giá trị của x
#Dùng phương pháp Gradient Descent với Learning rate alpha = 0.1 để cập nhật giá trị x trong vòng 10 vòng lặp

#BTVN 3:
# Tạo một tập dữ liệu giả lập với x là số giờ học ( Ngẫu nhiên từ 1 - 10) và y là số điểm
# trong cong thức y = 3x + 5 + noise . Với noise là một giá trị ngẫu nhiên nhỏ
# 1. Khởi tạo tham số w và b ngẫu nhiên với requires_grad = True
# 2. Tính MSE
# 3. Tính Gradient
# 4. Cập nhật tham số w và b bằng Gradient Descent với Learning rate alpha = 0.1
# 5. Lập lại quá trình trên trong 100 vòng lặp để xem sự hội tụ của mô hình
```

BTVN 1

```python
import torch
x = torch.tensor(2.0,requires_grad=True)
y = 5*x**5 + 6*x**3 -3*x + 1
print(y)
```

BTVN 2

```python
#Tạo một tensor ban đầu có giá trị là 2
# Learning rate và số vòng lặp
# Định nghĩa hàm
# Tính gradient
# Lưu kết quả
# Cập nhật x
# Reset gradient
# Tạo bảng kết quả
x = torch.tensor(2.0, requires_grad=True)
alpha = 0.1
epochs = 10
result = []
for i in range(epochs):
    y = x**3 + 2*x**2 + 5*x + 1
    y.backward()
    result.append([i+1, x.item(), y.item(), x.grad.item()])
    with torch.no_grad():
        x -= alpha * x.grad
    x.grad.zero_()

df = pd.DataFrame(result, columns=["Iteration", "x", "y", "dy/dx"])
print(df)
```

BTVN 3

```python
# Tạo dữ liệu
# Khởi tạo tham số
# Dự đoán
# MSE Loss
# Tính gradient
# Lưu kết quả
# Cập nhật Gradient Descent
# Reset gradient
np.random.seed(0)
x = torch.tensor(np.random.uniform(1, 10, 20), dtype=torch.float32)
noise = torch.randn(20) * 0.5
y = 3 * x + 5 + noise

w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

alpha = 0.1
epochs = 100
log = []

for i in range(epochs):
    y_pred = w * x + b
    loss = ((y_pred - y) ** 2).mean()
    loss.backward()
    log.append([i+1, loss.item(), w.item(), b.item()])
    with torch.no_grad():
        w -= alpha * w.grad
        b -= alpha * b.grad
    w.grad.zero_()
    b.grad.zero_()

df = pd.DataFrame(log, columns=["Iteration", "Loss", "w", "b"])
print(df.tail())
```
